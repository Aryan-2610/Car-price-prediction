```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

# Chrome options for Colab
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Correct ChromeDriver path for Colab
chrome_driver_path = "/usr/bin/chromedriver"

# Initialize driver
driver = webdriver.Chrome(service=Service(chrome_driver_path), options=chrome_options)

# URL
driver.get("https://www.quikr.com/cars/used+car+cars+delhi+x671")

# Scroll logic
last_number = driver.execute_script('return document.body.scrollHeight')

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(3)
    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == last_number:
        break
    last_number = new_height

# Extract HTML content
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')


# Close the driver
driver.quit()

```

```python
with open("page_content.html", "r", encoding="utf-8") as file:
    soup = BeautifulSoup(file, "html.parser")
```

```python
car_data = []

# Find all divs with class 'car-info__footer'
for car in soup.find_all('div', class_='car-info__footer'):
    car_name = car.find('h2').text.strip()
    features = car.find('div', class_='prime-features').text.strip()
    price = car.find('div', class_='price').text.strip()

    # Append data to list
    car_data.append([car_name, features, price])
```

```python
car_data
```

```python
import re
import numpy as np  # For np.nan

car_details = []

for egx in car_data:
    name = egx[0]
    company = name.split()[0]
    model = name.split()[1]
    year = name.split()[-1]

    features = egx[1]
    featureslist = features.split('/')

    kms = int(re.sub(r'[^0-9]', '', featureslist[0]))

    gas = featureslist[1].strip() if len(featureslist) > 1 else np.nan  # Fuel type as NaN for unknown

    if len(featureslist) > 2:
        owner_match = re.search(r'\d+', featureslist[2])
        owner = int(owner_match.group()) if owner_match else np.nan
    else:
        owner = np.nan  # Owner as NaN for unknown

    price = egx[2]
    price_int = int(re.sub(r'[^\d]', '', price))

    car_details.append([company, model, year, kms, gas, owner, price_int])

# Sample Output
for car in car_details[:5]:
    print(car)
```

```python
from sklearn.impute import SimpleImputer
import pandas as pd

# Create an imputer for categorical data using the mode
imputer = SimpleImputer(strategy='most_frequent')

# Fill missing values in 'Owner' column
df['Owner'] = imputer.fit_transform(df[['Owner']])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), ['Company', 'Model', 'Fuel']),
    remainder='passthrough'  # Keeps remaining columns as they are
)

# Linear regression model
lr = LinearRegression()

# Pipeline combining column transformer and model
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
scores = []

# Running the loop to try different random states
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    scores.append(r2_score(y_test, y_pred))

# Identify the best-performing split
best_random_state = np.argmax(scores)
best_r2_score = max(scores)
```

```python
print(best_random_state)
print(best_r2_score)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=522)
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test, y_pred))
```
